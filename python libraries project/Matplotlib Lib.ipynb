# Matplotlib Library

```python
import matplotlib.pyplot as plt
```

# Linear graph

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [4,5,6,7]
c = 'g'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
c = 'b'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
c = 'b'
plt.plot(x,y,c,marker ='o')
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.plot(y,marker = ">")
plt.plot(x,marker = 'o')
plt.show()
```

# Bar graph

```python
x = [1,3,6,9]
y = [5,7,10,13]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','b','y','g']
plt.bar(x,y,color = c)
plt.show()
```

```python
x = ["A","B","C","D"]
y = [4,2,7,9]
c = ['r','b','y','g']
plt.barh(x,y, color = c)
```

```python
x = ["A","B","C","D"]
y = [4,2,7,9]
plt.bar(x,y, color = "b")
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','b','y','g']
plt.barh(x,y,color = c)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','b','y','g']
plt.bar(x,y,color = c, width = 0.3)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','b','y','g']
plt.barh(x,y,color = c, height = 0.1)
plt.show()
```

 # Scatter plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,3,1]
plt.title("Scatter graph")
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,4,3]
plt.title("Scatter graph")
c = ["red","blue","green","yellow","magenta"]
plt.scatter(x,y, color = c, s = 200)
plt.xlabel("month")
plt.ylabel("no' of sale")
plt.show() 
```
