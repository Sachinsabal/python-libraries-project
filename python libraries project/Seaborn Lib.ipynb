# Barplot in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex")
```

## order

```python
order1 = ["Dream", "Biscoe","Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1)
plt.show()
```

## Hue order

```python
order1 = ["Dream", "Biscoe","Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1, hue_order = ["Female","Male"])
plt.show()
```

```python
order1 = ["Dream", "Biscoe","Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, color = "red")
```

# Histogram in seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df
```

```python
sns.displot(df["flipper_length_mm"])
plt.show()
```

# bins

```python
sns.displot(df["flipper_length_mm"], bins = [170,180,190,200,210,220,230])
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins = [170,180,190,200,210,220,230], color = "pink")
plt.show()
```

## kde

```python
sns.displot(df["flipper_length_mm"], bins = [170,180,190,200,210,220,230], color = "grey", kde = "True")
plt.show()
```
